```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("geyser")
# df = pd.read_csv('../Data/Otodom/apartaments.csv')
```

```python
df.head(5)
```

```python
sns.kdeplot(data=df, x='duration', y='waiting')
```

```python
sns.kdeplot(data=df, x='duration', y='waiting', hue='kind', palette='crest')
```

```python
sns.kdeplot(data=df, x='duration', y='waiting', hue='kind', levels=20)
```

```python
sns.kdeplot(data=df, x='duration', y='waiting', hue='kind', fill=True)
```

```python
sns.kdeplot(data=df, x='duration', y='waiting', hue='kind', thresh=.3)
```
